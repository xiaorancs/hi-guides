```python
lastActionTime = ['last1ActionTime','last2ActionTime','last3ActionTime', 
                  'last4ActionTime','last5ActionTime','last6ActionTime',
                  'last7ActionTime','last8ActionTime','last9ActionTime','last10ActionTime',]

typesActionTime = ['type1Last1Time','type234Last1Time','type56789Last1Time',
                   'type1Last2Time','type234Last2Time','type56789Last2Time',
                   'type1Last3Time','type234Last3Time','type56789Last3Time',
                   'type1Last4Time','type234Last4Time','type56789Last4Time',
                   'type1Last5Time','type234Last5Time','type56789Last5Time',]
typeActionTime = ['type2Last1Time','type3Last1Time','type4Last1Time','type5Last1Time', 
                  'type6Last1Time','type7Last1Time','type8Last1Time','type9Last1Time',
                  'type2Last2Time','type3Last2Time','type4Last2Time','type5Last2Time', 
                  'type6Last2Time','type7Last2Time','type8Last2Time','type9Last2Time',
                  'type2Last3Time','type3Last3Time','type4Last3Time','type5Last3Time', 
                  'type6Last3Time','type7Last3Time','type8Last3Time','type9Last3Time']
orderTime = 'last1OrderTime'
```

```python
df_train_11 = pd.DataFrame()
df_test_11 = pd.DataFrame()

df_train_11['userid'] = df_train['userid']
df_test_11['userid'] = df_test['userid']
```

```python
column = []

for f1 in lastActionTime[:6]:
    for f2 in typesActionTime[:9]:
        column.append(f1+'Sub'+f2)
        df_train_11[f1+'Sub'+f2] = df_train[f1] - df_train[f2]
        df_test_11[f1+'Sub'+f2] = df_test[f1] - df_test[f2]

for f1 in lastActionTime[:6]:
    for f2 in typeActionTime[:8]:
        column.append(f1+'Sub'+f2)
        df_train_11[f1+'Sub'+f2] = df_train[f1] - df_train[f2]
        df_test_11[f1+'Sub'+f2] = df_test[f1] - df_test[f2]


for f1 in typesActionTime[:9]:
    for f2 in typeActionTime[:8]:
        column.append(f1+'Sub'+f2)
        df_train_11[f1+'Sub'+f2] = df_train[f1] - df_train[f2]
        df_test_11[f1+'Sub'+f2] = df_test[f1] - df_test[f2]


for f in lastActionTime[:6]:
    column.append(f+'Sub'+orderTime)
    df_train_11[f+'Sub'+orderTime] = df_train[f] - df_train[orderTime]
    df_test_11[f+'Sub'+orderTime] = df_test[f] - df_test[orderTime]

for f in typesActionTime[:9]:
    column.append(f+'Sub'+orderTime)
    df_train_11[f+'Sub'+orderTime] = df_train[f] - df_train[orderTime]
    df_test_11[f+'Sub'+orderTime] = df_test[f] - df_test[orderTime]

for f in typeActionTime[:8]:
    column.append(f+'Sub'+orderTime)
    df_train_11[f+'Sub'+orderTime] = df_train[f] - df_train[orderTime]
    df_test_11[f+'Sub'+orderTime] = df_test[f] - df_test[orderTime]

len(column)
```

```python
df_train_11.to_csv('../data/dataSet/df_train_11.csv',index=False)
df_test_11.to_csv('../data/dataSet/df_test_11.csv',index=False)
```
