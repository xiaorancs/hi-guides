```python
import pandas as pd
import numpy as np
```

```python
# 采用加权平均的方法
sub1 = pd.read_csv('../result/sub_xgb_lr.csv') # lr 0.9621  1
sub2 = pd.read_csv('../result/subAvg.csv') # xgb 0.9647     3
sub3 = pd.read_csv('../result/sub_base.csv') # xgb 0.9639   1


sumall = 1 + 1 + 1
```

```python
sub1['orderType'] = sub1['orderType'] * 1 + sub2['orderType'] * 1 + sub3['orderType'] * 1

sub1['orderType'] = sub1['orderType'] / sumall
                     
```

```python
sub1.to_csv('../result/sub_Avg_xgblr_Avg_base.csv',index=False) 
```
